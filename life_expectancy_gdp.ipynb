# Setup the Countries to create the first line charts

```python
print(alldf.Country.unique())
chile = alldf[alldf.Country == 'Chile']
china = alldf[alldf.Country == 'China']
germany = alldf[alldf.Country == 'Germany']
mexico = alldf[alldf.Country == 'Mexico']
usa = alldf[alldf.Country == 'United States of America']
zimbabwe = alldf[alldf.Country == 'Zimbabwe']
#testing
print(chile)
```

# Create Line Charts to compare Life Expectancy and GDP by Country over the years

```python
plt.figure(figsize=(10, 8))
plt.plot(chile['Year'], chile['Life expectancy at birth (years)'], color='red')
plt.plot(china['Year'], china['Life expectancy at birth (years)'], color='yellow')
plt.plot(germany['Year'], germany['Life expectancy at birth (years)'], color='black')
plt.plot(mexico['Year'], mexico['Life expectancy at birth (years)'], color='green')
plt.plot(usa['Year'], usa['Life expectancy at birth (years)'], color='blue')
plt.plot(zimbabwe['Year'], zimbabwe['Life expectancy at birth (years)'], color='brown')
plt.title('Life Expectancy over the Years per Country')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.legend(['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe'],loc=4)
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(chile['Year'], chile['GDP'], color='red')
plt.plot(china['Year'], china['GDP'], color='yellow')
plt.plot(germany['Year'], germany['GDP'], color='black')
plt.plot(mexico['Year'], mexico['GDP'], color='green')
plt.plot(usa['Year'], usa['GDP'], color='blue')
plt.plot(zimbabwe['Year'], zimbabwe['GDP'], color='brown')
plt.title('GDP over the Years per Country')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend(['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe'],loc=4)
plt.show()
plt.clf()
```

# Create Scatterplots of Life Expectancy vs. GDP per Country over the years to see correlation.

```python
lifex = alldf['Life expectancy at birth (years)']
gdp = alldf['GDP']
country = alldf['Country']
print(lifex.head())
print(gdp.head())
print(country.unique())
plt.figure(figsize=(10, 8))
sns.scatterplot(x = lifex, y = gdp, hue = country)
plt.title('Life Expectancy vs. GDP per Country per Year')
plt.show()
plt.clf()
```

# Create Box Plots showing the average life expectancies per Country

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x = country, y = lifex)
plt.title('Average Life Expectancy Across Countries')
plt.show()
plt.clf()
```

# Create overlapping histograms to visualize the life expectancy distributions across the Countries

```python
plt.figure(figsize=(10, 8))
plt.hist(chile['Life expectancy at birth (years)'], label='Chile', density=True, alpha=.5)
plt.hist(china['Life expectancy at birth (years)'], label='China', density=True, alpha=.5)
plt.hist(germany['Life expectancy at birth (years)'], label='Germany', density=True, alpha=.5)
plt.hist(mexico['Life expectancy at birth (years)'], label='Mexico', density=True, alpha=.5)
plt.hist(usa['Life expectancy at birth (years)'], label='USA', density=True, alpha=.5)
plt.hist(zimbabwe['Life expectancy at birth (years)'], label='Zimbabwe', density=True, alpha=.5)
plt.title('Life Expectancy Distributions across Countries')
plt.legend()
plt.show()
plt.clf()
```

# Create separate histogramsto visualize the life expectancy distributions across the Countries¶

```python
plt.subplot(3,2,1)
plt.hist(chile['Life expectancy at birth (years)'], color='red')
plt.title('Chile Life Expectancy Distribution')

plt.subplot(3,2,2)
plt.hist(china['Life expectancy at birth (years)'], color='yellow')
plt.title('China Life Expectancy Distribution')

plt.subplot(3,2,3)
plt.hist(germany['Life expectancy at birth (years)'], color='black')
plt.title('Germany Life Expectancy Distribution')

plt.subplot(3,2,4)
plt.hist(mexico['Life expectancy at birth (years)'], color='green')
plt.title('Mexico Life Expectancy Distribution')

plt.subplot(3,2,5)
plt.hist(usa['Life expectancy at birth (years)'], color='blue')
plt.title('USA Life Expectancy Distribution')

plt.subplot(3,2,6)
plt.hist(zimbabwe['Life expectancy at birth (years)'], color='brown')
plt.title('Zimbabwe Life Expectancy Distribution')

plt.subplots_adjust(wspace=0.9, hspace=0.7)
plt.show()
plt.clf()
```
